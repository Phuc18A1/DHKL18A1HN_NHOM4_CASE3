# 1. Đọc campaign_A.csv và campaign_B.csv bằng np.loadtxt hoặc np.genfromtxt.

```python
import numpy as np
data1=np.loadtxt('campaign_A.csv',delimiter=',',skiprows=1,dtype=str)
print('đọc file campaign_A.csv')
print(data1)

data2=np.loadtxt('campaign_B.csv',delimiter=',',skiprows=1,dtype=str)
print('\nđọc file campaign_B.csv')
print(data2)
```

# 2. Tách cột ngày và cột lượng truy cập thành hai mảng riêng

```python
ngày_a=data1[:,0]
số_lượng_truy_cập_a=data1[:,1].astype(int)
print('cột ngày của chiến dich a')
print(ngày_a)
print('\n cột lượng truy cập của chiến dịch a')
print(số_lượng_truy_cập_a)

ngày_b=data2[:,0]
số_lượng_truy_cập_b=data2[:,1].astype(int)
print('\ncột ngày của chiến dịch b')
print(ngày_b)
print('\n cột lượng truy cập của chiến dịch b')
print(số_lượng_truy_cập_b)
```

# 3. Chuyển cột ngày sang định dạng datetime64[D].

```python
ngày_a=data1[:,0].astype('datetime64[D]')
print('chuyển đổi cho chiến dịch a')
print(ngày_a)

ngày_b=data2[:,0].astype('datetime64[D]')
print('\nchuyển đổi cho chiến dịch b')
print(ngày_b)
```

# 6.Boolean mask: tìm các ngày Campaign A có lượng truy cập cao hơn Campaign B.

```python
số_lượng_truy_cập_b=data2[:,1].astype(int)
mask=số_lượng_truy_cập_a>số_lượng_truy_cập_b
ngày_có_số_lượng_cao_hơn=ngày_a[mask]
số_lượng_truy_cập_a_cao_hơn_b=số_lượng_truy_cập_a[mask]
số_lượng_truy_cập_b_thấp_hơn_a=số_lượng_truy_cập_b[mask]
print('các ngày campaign A có số lượng truy cập cao hơn campaign B:')
for d,a,b in zip(ngày_có_số_lượng_cao_hơn,số_lượng_truy_cập_a_cao_hơn_b,số_lượng_truy_cập_b_thấp_hơn_a):
    print(f'{d}: a={a},b={b}')
```

# 7.Boolean mask: tìm các ngày Campaign B tăng trưởng đột biến so với ngày trước đó.

```python
tăng_trưởng=số_lượng_truy_cập_b[1:]-số_lượng_truy_cập_b[:-1]
mask=tăng_trưởng>2000
ngày_tăng_đột_biến=ngày_b[1:][mask]
số_lượng_truy_cập_b_cao_đột_biến=số_lượng_truy_cập_b[1:][mask]
số_lượng_truy_cập_b_trước_đó=số_lượng_truy_cập_b[:-1][mask]
print('các ngày campaign b có mức tăng trưởng đột biến so với hôm trc đó:')
for d,a,b in zip(ngày_tăng_đột_biến,số_lượng_truy_cập_b_cao_đột_biến,số_lượng_truy_cập_b_trước_đó):
    print(f'{d}: hôm trc={b}, hôm nay={a},tăng={a-b}')
```

# 8. Fancy indexing: tìm top-5 ngày có lượng truy cập cao nhất của mỗi chiến dịch.

```python
sap_xep_a=np.argsort(số_lượng_truy_cập_a)[::-1]
sap_xep_b=np.argsort(số_lượng_truy_cập_b)[::-1]
top_5_ngày_có_số_lượng_nhiều_nhất_của_a=ngày_a[sap_xep_a[:5]]
top_5_số_lượng_nhiều_nhất_của_a=số_lượng_truy_cập_a[sap_xep_a[:5]]
top_5_ngày_có_số_lượng_nhiều_nhất_của_b=ngày_b[sap_xep_b[:5]]
top_5_số_lượng_nhiều_nhất_của_b=số_lượng_truy_cập_b[sap_xep_b[:5]]
print('top 5 ngày chiến dịch a có số lượng nhiều nhất: ')
for a,b in zip(top_5_ngày_có_số_lượng_nhiều_nhất_của_a,top_5_số_lượng_nhiều_nhất_của_a):
    print(f'{a}:{b}')
print('\ntop 5 ngày chiến dịch b có số lượng nhiều nhất: ')
for a,b in zip(top_5_ngày_có_số_lượng_nhiều_nhất_của_b,top_5_số_lượng_nhiều_nhất_của_b):
   print(f'{a}:{b}')
```

# 9. Broadcasting: tính chênh lệch truy cập từng ngày giữa hai chiến dịch.

```python
chênh_lệch=số_lượng_truy_cập_a-số_lượng_truy_cập_b
print('\nchênh lệch truy cập từng ngày giữa 2 chiến dịch: ')
for a,b in zip(ngày_a,chênh_lệch):
    print(f'{a}:A-B= {b:+}')
```

# 10. Dùng np.diff() để phân tích biến động truy cập ngày liền kề của từng chiến dịch.

```python
biến_động_a=np.diff(số_lượng_truy_cập_a)
biến_động_b=np.diff(số_lượng_truy_cập_b)
print('\nbiến động truy cập ngày liền kề của chiến dịch a: ')
for a,b,c in zip(ngày_a[1:],biến_động_a,ngày_a[0:]):
    print(f'{a} so với {c}: thay đổi {b:+}') 
print('\nbiến động truy cập ngày liền kề của chiến dịch b: ')
for a,b,c in zip(ngày_b[1:],biến_động_b,ngày_b[0:]):
    print(f'{a} so với {c}: thay đổi {b:+}')
```
